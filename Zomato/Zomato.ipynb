```python
#import important libraries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Error:utf-8' codec can't decode byte 0xed in position 7044: invalid continuation byte
# need to use encoding
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head() #first five rows will be displayed
```

UTF-8 and Latin-1 encoding are used for electronic communication. UTF-8 is utilized to convert alphanumeric data into machine language for processing. At a granular level, machine language is represented in 0s and 1s. Therefore, encoding is necessary to convert incoming English language alphabets into machine language for processing by a machine.

If the incoming language contains characters that are not defined in the English language, the machine will be unable to convert it into 0s and 1s, as it is considered a foreign word or character not listed in the UTF-8 reference directory. In such cases, appropriate references such as JIS encoding for Japanese characters or Latin-1 for Latin characters need to be provided.

```python
# read_excel as it is a excel file
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
Ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
Ratings
```

* df.size() : returns a single integer
*   df.groupby.size(): It returns a pandas series that possess the total number of row count for each group

*   df.reset_index(): reset the index of the DataFrame, and use the default one instead When we reset the index, the old index is added as a column, and a new sequential index is used

*   df.rename(rows={0:'name'}): Rename index labels

*   df.rename(columns={0:'name'}): Rename columns labels






```python
Ratings=Ratings.reset_index().rename(columns={0:'Rating Count'})
Ratings
```

Ratings
*   4.5 - 4.9 indicates Excellent -> Dark Green
*   4.0 - 4.5 indicates Very Good -> Green
*   3.5 - 3.9 indicates  Good -> Yellow
*   2.5 - 3.4 indicates Average -> Orange
*   1.8 - 2.4 indicates Poor -> Red
*   0.0 indicates Not Rated -> White





```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color', data=Ratings,palette=['lightgray','red','orange','yellow','lightgreen','green'])
```

```python
Total_ratings=Ratings["Rating Count"].sum()
Total_ratings
```

```python
Ratings_mean=Ratings["Aggregate rating"].mean()
Ratings_median=Ratings["Aggregate rating"].median()
print("Mean Rating is : ",Ratings_mean)
print("Median Rating is : ",Ratings_median)
```

```python
(2148/Total_ratings)*100
```

```python
# count plot is used to represent the occurrence(counts) of the observation present in the categorical variable
sns.countplot(x='Rating color',data=Ratings,palette=['lightgray','red','orange','yellow','lightgreen','green'])
```

Observation
* Not rated count is very high around 22.48% did not rate
* Maximum people rated between 2.5 to 3.4 according to count and bar plot
* Mean and median rating is 3.24 and 3.3 respectively
