```python
import pandas as pd
DF=pd.read_csv("covid-19 dataset.csv")
print(DF)
```

# **Ques1**
Statewise Male vs Female vaccinated ratio on a specific date

```python
import matplotlib.pyplot as plt
x=str(input("Enter the date(dd-mm-yy)"))
df=DF.loc[DF["Date (Vaccine)"]==x,["State","Male","Female"]]
df.plot(kind="line",color=["orange","green"],marker="*",markersize=2,linewidth=1)
plt.title("Male vs Female vaccinated")
plt.xlabel("<----State---->")
plt.ylabel("No. of Male and Female-->")
ticks=df.index.tolist()
plt.xticks(ticks,df.State,rotation=90)
plt.show()
```

# **Ques2**
Statewise Death and Cured people on a specific date

```python
import matplotlib.pyplot as plt
x=str(input("Enter the date(dd-mm-yy)"))
df=DF.loc[DF["Date"]==x,["State/UnionTerritory","Deaths","Cured"]]
df.plot(kind="line",color=["crimson","yellow"],marker="*",markersize=2,linewidth=1)
plt.title("  Death and Cured registered ")
plt.xlabel("<----State---->")
plt.ylabel("<---Death vs Cured-->")
ticks=df.index.tolist()
plt.xticks(ticks,df["State/UnionTerritory"],rotation=90)
plt.show()
```

# **Ques3**

# **Ques5**
Total Confirmed cases statewise

```python
import seaborn as sns
import plotly
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(28, 28))
data = DF[['State/UnionTerritory','Confirmed']]
sns.barplot(x="Confirmed",y="State/UnionTerritory",data=data,label="Total", color="plum",ci=None)
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 700000), ylabel="States",xlabel="Cases")
plt.show()
```

# **Ques6**

```python
import seaborn as sns
import plotly
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(28, 28))
data = DF[['State/UnionTerritory','Deaths']]
sns.barplot(x="Deaths",y="State/UnionTerritory",data=data, label="Cured", color="orange",ci=None)
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 20000), ylabel="States",xlabel="Cases")
plt.show()
```

# **Ques7**
Analysis on fatality ratio among states

```python

DF['Fatality-Ratio'] = DF['Deaths']/DF['Confirmed']
df1=DF.groupby('State/UnionTerritory')['Fatality-Ratio'].sum().reset_index()
df1.plot(kind='line',x='State/UnionTerritory',y='Fatality-Ratio',title='Fatality-Ratio Among States',color='lime')
ticks=df1.index.tolist()
plt.xticks(ticks,df1["State/UnionTerritory"],rotation=90)
plt.show()
```

# **Ques13**
Total confirmed recovered and death cases overall

```python
df_confirmed_india=DF.groupby('Date')['Confirmed'].sum().reset_index()
df_cured_india=DF.groupby('Date')['Cured'].sum().reset_index()
df_death_india=DF.groupby('Date')['Deaths'].sum().reset_index()
print("The Confirmed Cases are",df_confirmed_india.Confirmed.max())
print("The Recovered Cases are",df_cured_india.Cured.max())
print("The Deaths Cases are",df_death_india.Deaths.max())
```
